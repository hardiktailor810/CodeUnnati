```python
#import matplotlib.pyplot
import matplotlib.pyplot as plt
import numpy as np

#draw plot as fig
fig, ax = plt.subplots()

x = np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis('equal')

#text annotation to the plot where it indicate maximum  values of the curv
ax.annotate('local maximum', xy = (6.28,1), xytext = (10,4),
            arrowprops=dict(facecolor='black', shrink=0.04))

#text annotation to the plot where it indicate maximum  values of the curv
ax.annotate('local minimum', xy = (5 * np.pi, -1), xytext = (2,-6),
            arrowprops=dict(arrowstyle="->",connectionstyle="angle3,angleA=0,angleB=-90"));
```

# Pie Chart

Pie chart is circular staastical plot that can display only one series of data.

The area of the chart is the total % of the given data  

```python
#import libraries

import numpy as np
import matplotlib.pyplot as plt

#creating datset
Branch = ["Computer", "Electrical" , "Mechanical" , "Civil" , "Automobile" , "IT"]

data = [50, 34, 40, 29, 12, 23]

#create chart

fig = plt.figure(figsize = (10,7))
plt.pie(data, labels = Branch)
plt.show()
```

```python
#import libraries

import numpy as np
import matplotlib.pyplot as plt

#creating datset
Branch = ["Computer", "Electrical" , "Mechanical" , "Civil" , "Automobile" , "IT"]

data = [50, 34, 40, 29, 12, 23]
explode = (0.2 , 0 , 0 , 0 , 0 , 0)

#create chart

fig = plt.figure(figsize = (10,7))
plt.pie(data, labels = Branch , explode = explode)
plt.show()



```

# 3D Ploting

```python
#import mplot3d toolkit ,  numpy & matplotlib

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d #3d toolkit lib

```

```python
fig = plt.figure()

#syntex for 3D projection

ax = plt.axes(projection = '3d')

#defining all 3 axes

z = np.linspace(0 , 1 , 100)
x = z * np.sin(25*z)
y = z * np.cos(25*z)

#3d ploting

ax.plot3D(x , y , z , 'purple')
ax.set_title('3D Plotimg')
plt.show()
```
